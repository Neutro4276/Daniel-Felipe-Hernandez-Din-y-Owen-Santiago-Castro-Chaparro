```python

# ejercicio 12
VALOR_HORA = 5000
PORCENTAJE_RETENCION = 0.123  # Representa el 12.3%


horas_trabajadas = float(input("Ingrese la cantidad de horas trabajadas en la semana: "))

salario_bruto = horas_trabajadas * VALOR_HORA
retencion = salario_bruto * PORCENTAJE_RETENCION
salario_neto = salario_bruto - retencion


print("-" * 30)
print(f"RESUMEN DE PAGO:")
print(f"Salario Bruto:       ${salario_bruto:,.2f}")
print(f"Retención (12.3%):   ${retencion:,.2f}")
print(f"Salario Neto:        ${salario_neto:,.2f}")
print("-" * 30)
```
